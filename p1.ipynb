```python
import numpy as np
np.random.seed(0)

def reciprocal_values(n):
    output = np.empty(len(n))
    for i in range(len(n)):
        output[i] = 1/n[i]
    return output    
            
n=np.random.randint(1,10,size = 10)
reciprocal_values(n)
    
    
```
